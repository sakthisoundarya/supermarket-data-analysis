```python
#thendral supermarket highest gross income branch

branch_gross_income=sales_df.groupby('Branch')['gross income'].sum().reset_index()

plt.figure(figsize=(4, 5))
sns.barplot(x='Branch', y='gross income', data=branch_gross_income, palette='viridis')
plt.title('Total Gross Income by Branch')
plt.xlabel('Branch')
plt.ylabel('Total Gross Income')
plt.show()
```

```python
#lets see the highest revenue branch


branch_revenue=sales_df.groupby('Branch')['Total'].sum().reset_index()


plt.figure(figsize=(4, 5))
sns.barplot(x='Branch' , y='Total' , data=branch_revenue, palette='viridis')
plt.title('Total Gross Income by Branch')
plt.xlabel('Branch')
plt.ylabel('Total gross income')
plt.show()
```

```python
#lets see which payment mode is mst preferred by users

payment_counts=sales_df['Payment'].value_counts().reset_index()
payment_counts.columns=['Payment' , 'count']


plt.figure(figsize=(4, 5))
sns.barplot(x='Payment' , y='count',data=payment_counts, palette='pastel')

plt.title('payment modes and counts')
plt.xlabel('payment mode')
plt.ylabel('count')
plt.show()

```

It seems people prefer Ewallet and cash as favourite and convinient modes of payment.so,May be there's a opportunity fot the company to collabrate with credit card companies to offer discounts which would lead to more sales.

```python
#lets see which gender constitutes highest revenue

gender_revenue=sales_df.groupby('Gender')['Total'].sum().reset_index()
plt.figure(figsize=(4, 5))
sns.barplot(x='Gender' , y='Total',data=gender_revenue, palette='viridis')
plt.title('Revenue by Gender')
plt.xlabel('Gender')
plt.ylabel('Total')
plt.show()
```

```python
#lets see the distribution of product counts lines

product_line_counts=sales_df['Product line'].value_counts().reset_index()
product_line_counts.columns= ['Product line', 'Count']

plt.figure(figsize=(10, 6))
barplot=sns.barplot(x='Product line', y='Count',data=product_line_counts, palette='pastel')
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title('product lines and counts')
plt.xlabel('Product lines')
plt.ylabel('Counts')
plt.show()
```

```python
#lets see which product line constitutes to highest product revenue



product_line_revenue=sales_df.groupby('Product line')['Total'].sum().reset_index()
plt.figure(figsize=(8, 6))
barplot=sns.barplot(x='Product line', y='Total', data=product_line_revenue, palette='viridis')
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title('Revenue by Product line')
plt.xlabel('Product line')
plt.ylabel('Total revenue')
plt.show()

```

As compared to the product sales fashion accessories has more sale that food and beverages but the revenue of the supermarket was increased by food and beverages,to increase the sale of food and beverages it is advisable to provide more offers in this product line which will increase the supermarket revenue as well

```python
sales_df['Date'] = pd.to_datetime(sales_df['Date'], errors='coerce')

# Drop rows with missing values in 'Date' or 'Total'
sales_df = sales_df.dropna(subset=['Date', 'Total'])

# Extract day of the week from 'Date' and create a new column 'DayOfWeek'
sales_df['DayOfWeek'] = sales_df['Date'].dt.day_name()

# Aggregate total sales for each day of the week
daywise_sales = sales_df.groupby('DayOfWeek')['Total'].sum().reset_index()

# Sort days of the week
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daywise_sales['DayOfWeek'] = pd.Categorical(daywise_sales['DayOfWeek'], categories=days_order, ordered=True)
daywise_sales = daywise_sales.sort_values('DayOfWeek')

# Create a bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='DayOfWeek', y='Total', data=daywise_sales, palette='viridis')

# Customize the plot
plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')

plt.show()
```

so saturday contributed more sales followed by tuesday,so the supermarket would be more busy on that days,so it is advisable to add more staffs and try to provide offer day sale on other days to attract more people on other days as well.
